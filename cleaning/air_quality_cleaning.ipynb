```python
import json
import csv
import pandas
```

```python
# Read sample data
with open('../data_raw/airquality/2003.json') as json_file: 
    data = json.load(json_file) 
  
airquality_data = data['Data'] 
```

```python
# Convert json to dataframe
df = pandas.read_json(json.dumps(airquality_data))
```

```python
df.describe()
```

```python
df.head(5)
```

```python
# Filter to the desired columns
df_filter = df.filter(['latitude', 'longitude', 'date_local', 'parameter', 'county', 'city', 'aqi', 'observation_count'], axis=1)
```

```python
df_filter.head(10)
```

```python
# Start with the 2003 data and concat all other years
df_concat = df_filter

for year in range(2004,2016):
    with open('../data_raw/airquality/' + str(year) + '.json') as json_file: 
        data = json.load(json_file) 
    df = pandas.read_json(json.dumps(data['Data']))
    df_filter = df.filter(['latitude', 'longitude', 'date_local', 'parameter', 'county', 'city', 'aqi', 'observation_count'], axis=1)
    df_concat = pandas.concat([df_concat, df_filter], ignore_index=True)
```

```python
df_concat.describe()
```

```python
# Write to csv
df_concat.to_csv('../data/airquality.csv', index=False)
```
